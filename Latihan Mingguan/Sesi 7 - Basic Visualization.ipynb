### Two Types of Plotting

```python
#opsi 1

df_top5.plot(kind='area',
            alpha=0.35,
            figsize=(20,10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
#opsi 2

ax=df_top5.plot(kind='area',
            alpha=0.35,
            figsize=(20,10),
            )
ax.set_title('Immigration Trend of 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

### Pie Charts

```python
df_continents=df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))

df_continents.head()
```

### Bubble Plots

```python
df_can_t=df_can[years].transpose()
df_can_t.index=map(int, df_can_t.index)
df_can_t.index.name='Year'
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
norm_USA=(df_can_t['United States of America'] - df_can_t['United States of America'].min())/(df_can_t['United States of America'].max() - df_can_t['United States of America'].min())

norm_UK=(df_can_t['United Kingdom of Great Britain and Northern Ireland'] - df_can_t['United Kingdom of Great Britain and Northern Ireland'].min())/(df_can_t['United Kingdom of Great Britain and Northern Ireland'].max() - df_can_t['United Kingdom of Great Britain and Northern Ireland'].min())
```

```python
ax0=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='United States of America',
                 figsize=(14,8),
                 alpha=0.5,
                 color='firebrick',
                 s=norm_USA * 2000 + 10,
                 xlim=(1975,2015)
                 )

ax1=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='United Kingdom of Great Britain and Northern Ireland',
                 alpha=0.5,
                 color='midnightblue',
                 s=norm_UK * 2000 + 10,
                 ax=ax0
                 )

ax0.set_title('Immigration from United States of America and United Kingdom of Great Britain and Northern Ireland from 1980 - 2013')
ax0.set_ylabel('Number of immigrants')
ax0.legend(['United States of America','United Kingdom of Great Britain and Northern Ireland'], loc='upper left', fontsize='x-large')
```

```python
norm_JPN=(df_can_t['Japan'] - df_can_t['Japan'].min())/(df_can_t['Japan'].max() - df_can_t['Japan'].min())

norm_SIN=(df_can_t['Singapore'] - df_can_t['Singapore'].min())/(df_can_t['Singapore'].max() - df_can_t['Singapore'].min())
```

```python
ax0=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='Japan',
                 figsize=(14,8),
                 alpha=0.5,
                 color='magenta',
                 s=norm_JPN * 1300 + 10,
                 xlim=(1975,2015)
                 )

ax1=df_can_t.plot(kind='scatter',
                 x='Year',
                 y='Singapore',
                 alpha=0.5,
                 color='blue',
                 s=norm_SIN * 1300 + 10,
                 ax=ax0
                 )

ax0.set_title('Immigration from Japan and Singapore from 1980 - 2013')
ax0.set_ylabel('Number of immigrants')
ax0.legend(['Japan','Singapore'], loc='upper left', fontsize='x-large')
```
